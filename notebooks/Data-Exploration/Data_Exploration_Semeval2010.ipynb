```python
import os, pandas as pd, numpy as np

RESOURCE_PATH = "/data/medg/misc/semeval_2010"
def res(path): return os.path.join(RESOURCE_PATH, path)
import math
```

```python
dictionary = {0:'Component-Whole(e2,e1)', 1:'Instrument-Agency(e2,e1)', 2:'Member-Collection(e1,e2)', 
3:'Cause-Effect(e2,e1)', 4:'Entity-Destination(e1,e2)', 5:'Content-Container(e1,e2)', 
6:'Message-Topic(e1,e2)', 7:'Product-Producer(e2,e1)', 8:'Member-Collection(e2,e1)', 
9:'Entity-Origin(e1,e2)', 10:'Cause-Effect(e1,e2)', 11:'Component-Whole(e1,e2)', 
12:'Message-Topic(e2,e1)', 13:'Product-Producer(e1,e2)', 14:'Entity-Origin(e2,e1)', 
15:'Content-Container(e2,e1)', 16:'Instrument-Agency(e1,e2)', 17:'Entity-Destination(e2,e1)',
18:'Other'}
```

In the parsed data: 
Num1 num2 num3 num4 num5  sentence

Num1 - relation number

Num2 - left entity start (starts the numbering from 0)

Num3 - left entity end

Num4 - right entity start

Num5 - right entity end


```python
def entities_givendistance(filename, distance, toprint):
    needed_count = 0
    total_count = 0
    with open(res(filename)) as file:
        for line in file:
            total_count += 1
            line = line.split()
            relation = int(line[0])
            left_start = int(line[1])
            left_end = int(line[2])
            right_start = int(line[3])
            right_end = int(line[4])
            sentence = " ".join(str(x) for x in line[5:])
            list_sentence = line[5:]
            left = " ".join(str(x) for x in list_sentence[left_start:left_end+1])
            right = " ".join(str(x) for x in list_sentence[right_start:right_end+1])

            if(math.fabs(left_end - right_start) == distance):
                needed_count += 1
                if(toprint is True):
                    print(
                        'Left: {}\n'
                        'Right: {}\n'
                        'Label: {}\n'
                        'Sentence: {}\n'\
                        .format(left, right, dictionary[relation], sentence)
                    )
    return (float(needed_count)/total_count) * 100
```

# Entities with various distances

```python
distribution_of_entity_distance = pd.DataFrame(columns=['Train Data', 'Test Data'])
```

```python
distance = 1
while(True):
    percent_train = entities_givendistance('train.txt',distance, False)
    percent_test = entities_givendistance('test.txt',distance, False)
    percent_train = float("{0:.2f}".format(percent_train))
    percent_test = float("{0:.2f}".format(percent_test))
    distribution_of_entity_distance.loc[distance] = [percent_train, percent_test]
    if(percent_train == 0 and percent_test == 0):
        break;
    # index corresponds to the distance between the entities
    distance += 1
```

```python
distribution_of_entity_distance["Word Distance"] = distribution_of_entity_distance.index
distribution_of_entity_distance.reset_index(drop=True, inplace=True)
distribution_of_entity_distance
```

```python
# Set up spaCy
from spacy.en import English
from spacy.symbols import ORTH, LEMMA, POS
import re
parser = English()
import nltk
nltk.download('wordnet')
from nltk import wordnet as wn
import numpy as np

#input must be a tokenized sentence in the form ['There', 'is', 'a', 'dog', '.']
def get_hypernyms(tokenizedSentence):
    hypernyms = []
    for word in tokenizedSentence:
        hypernym_perword = []
        if len(wn.wordnet.synsets(word)) == 0:
            hypernyms.append("")
            continue
        foundhypernym = 0
        for synset in wn.wordnet.synsets(word):
            if len(synset.hypernyms()) > 0:
                hypernym = synset.hypernyms()[0]
                foundhypernym += 1
                hypernyms.append(hypernym.name())
                break
        if foundhypernym == 0:
            hypernyms.append("")
    return hypernyms
```

```python
def entities_hypernyms(filename, numExamples):
    #needed_count = 0
    total_count = 0
    with open(res(filename)) as file:
        for line in file:
            total_count += 1
            if(total_count > numExamples):
                return
            line = line.split()
            relation = int(line[0])
            left_start = int(line[1])
            left_end = int(line[2])
            right_start = int(line[3])
            right_end = int(line[4])
            sentence = " ".join(str(x) for x in line[5:])
            list_sentence = line[5:]
            left = " ".join(str(x) for x in list_sentence[left_start:left_end+1])
            right = " ".join(str(x) for x in list_sentence[right_start:right_end+1])
            left_list = left.split()
            right_list = right.split()
            left_hypernym_list = get_hypernyms(left_list)
            right_hypernym_list = get_hypernyms(right_list)
            left_hypernym = " ".join(map(str, left_hypernym_list))
            right_hypernym = " ".join(map(str, right_hypernym_list))
            #if(toprint is True):
            print(
                'Left: {}\t Hypernym: {}\n'
                'Right: {}\t Hypernym: {}\n'
                'Label: {}\n'
                'Sentence: {}\n'\
                .format(left, left_hypernym, right, right_hypernym, dictionary[relation], sentence)
            )
```
